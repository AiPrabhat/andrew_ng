```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math, copy
```

```python
# create target data
x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1, 1)
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, alpha, num_iters, cost_function, compute_gradient):
    J_history = []
    p_history = []
    b = 0
    w = 0
    
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(x, y, w, b)
        
        b = b - alpha * dj_db
        w = w - alpha * dj_dw

    return w, b
```

```python
iterations = 100000
tmp_alpha = 1.0e-2

w_final, b_final = gradient_descent(X, y, tmp_alpha, iterations, compute_cost, compute_gradient)

print(f"(w,b) found by gradient descent: ({w_final}, {b_final})")
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x,X*w_final + b_final, label="Predicted Value")
plt.title("no feature engineering")
plt.xlabel("X"); plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x**2

X = np.c_[x, x**2, x**3]
```

```python
iterations = 100000
tmp_alpha = 1.0e-7

w, b = gradient_descent(X, y, tmp_alpha, iterations, compute_cost, compute_gradient)

print(f"(w,b) found by gradient descent: ({w}, {b})")
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, X*w + b, label="Predicted Value")
plt.title("feature engineered")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
