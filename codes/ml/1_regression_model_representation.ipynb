# Regression Model Representation

```python
import numpy as np
import matplotlib.pyplot as plt
```

## **f<sub>w,b</sub>(x<sup>[i]</sup>) = wx<sup>[i]</sup> + b**

```python
x_train = np.array([1.0, 2.0, 3.0])      # This is x
y_train = np.array([300.0, 500.0, 700.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

### Number of training examples `m`

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of traning examples is: {m}")
```

### `Training example x_i, y_i`

```python
i = 0      # Change this to 1 to see (x^1, y^1)
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{i}), y^{i})) = ({x_i}, {y_i})")
```

### Plotting the data

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel("Price (in 1000's of dollars)")
# Set the x-axis label
plt.xlabel("Size (1000's of sqft)")
plt.show()
```

### **f<sub>w,b</sub>** = **f_wb**

**f_wb[i] = wx[i] + b**

```python
w = 200
b = 100

def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label="Our Prediction")

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label="Actual Values")

# Set the title
plt.title("Housing Prices")

# Set the y-axis label
plt.ylabel("Price (in 1000's of dollars)")

# Set the x-axis label
plt.xlabel("Size (in 1000's of sqft)")
plt.legend()
plt.show()
```

### Prediction
`Now that we have a model, we can use it to make our original prediction. Let's predict the price of a house with 1200 sqft. Since the units of $x$ are in 1000's of sqft, $x$ is 1.2.`

```python
w = 200
b = 100
x_i = 1.2

cost_1200sqft = w * x_i + b

print(f"${cost_1200sqft: .0f} thousand dollars")
```
